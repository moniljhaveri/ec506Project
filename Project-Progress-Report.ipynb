### Introduction

Y2Y Harvard Square is a youth homeless shelter located in Cambridge, MA. It is one of only a handful of homeless shelters that track youth homelessness and services guests between the ages of 18 to 25. This segment of the homeless population is amongst the least studied homeless population. This lack of academic and government research hinders attempts for positive outcomes. 

Y2Y provides a comforting escape from the streets and provides accommodations, amenities and resources to assist guests in establishing an adequate foundation. Guests can either stay for one night or 30-days. This is determined through a lottery that is held every night slightly after mandatory check-in time. Those who stay are provided with resources, such as Youth Workers who build relationships and mentor guests, Case Managers who hold workshops and assist with goal achievement, students from Harvard Law who provide legal assistance, and physicians who provide medical care. 

The lack of research in this space  makes understanding the data that Y2Y collects vitally important. The goal of this project is to use this data to explore the factors that best predict guest satisfaction scores and guest outcomes.

### Datasets

The datasets utilized for the project consist of information recorded at the Y2Y facility by the Y2Y staff members and consist of guest surveys,  attendance, student case manager notes, youth worker notes, visiting professionals’ notes, such as physicians and Harvard Law students, and job/housing outcomes. That data provides insight on each guest during their stay. The datasets do not hold a large number of guests, since the program has only run for two full seasons, but each guest has a substantial amount of information recorded, yielding a large number of features to be associated for analysis. Each dataset and how we are using the data is described in detail below.

**Guest Survey** 

The Guest Survey dataset consists of survey results given to each 30-day/long-term stay guest upon their first or second night at Y2Y. Each survey includes demographic information, such as age, sex, and race, as well as other information, such as education history, sexual orientation, housing situation and job status. This information is crucial for understanding Y2Y’s guests and providing them with statistical inferences. The project utilizes this data to not only run statistical analysis but also to determine the prevalent features for predicting a guest’s satisfaction rating after their stay and whether they were able to find housing or a job. Ths data is extracted and combined with the exit survey data and cross checked with the other datasets to establish features for statistical analysis and classifier prediction. 

**Exit Survey**

The exit survey contains each 30-day/long-term guests’ responses prior to their leave. This dataset includes information like guest satisfaction scores, comments about future plans, and whether the guests rate themselves in a better situation after their stay at Y2Y. This information is the baseline for determining how Y2Y is performing and if Y2Y is making a substantial difference in their guests’ lives. Statistical analysis is done on this data to give Y2Y a clearer understanding of how they are rated. The data is also used in combination with features from the guest surveys and other datasets, to train the classifiers to how satisfied a guest will be based on their responses to the the guest survey and their attendance. 

**Attendance**

The attendance dataset is a list of each guest's’ attendance status for every night and whether they were present, called in, entered late, won an emergency 1-night lottery bed, or did not show up. This information is used to correlate guests’ presence to satisfactory and  consequences. 

**Lottery**

The Lottery dataset consists of the participants and outcomes of the daily lottery. This includes who entered, when they entered, whether they entered for a 1-night stay or a 30-day stay and if they won or lost. The data is used as an indicator for recurrence or 1-night or 30-day stay guests and correlated to the satisfaction and consequences data.  

**Consequences**

The Consequences dataset set lists the consequences recorded. The information includes the guest involved, the type of consequence such as rules broken, the severity, and the sanction. This data is used for correlation analysis with the Lottery dataset and satisfactory rating. 

**Season 2 Guest Housing & Jobs**

The Guest Housing & Jobs dataset includes the housing and employment status of some of the guests during season 2 of Y2Y’s opening. This includes whether they found a job or housing sometime during their stay or upon leaving. The information is used as ground truth indicators for determining if guests leave homelessness or unemployment during or after their stay at Y2Y.  

**Youth Worker Notes & Workshops**

Notes that caseworkers and also includes notes about workshop attendance.

```python
def hist_exit_scores(df, column_score_names):
    for column in column_score_names:
        scores = df[column].dropna()
        print(column, "\n", "\n", df[column].describe(), "\n")
        bins=np.linspace(0, 9, 9, endpoint=False)
        # groups = resturaunt_rev_df.groupby(np.digitize(resturaunt_rev_df.stars, bins))
        fig, ax = plt.subplots(1,1, figsize=(12,9))
        ax.hist(scores, bins=bins, align='left', color="blue", rwidth=0.7)
        ax.set_xticks(bins)
        plt.title(column, size=16)
        plt.xlabel("Rating", size=16)
        plt.ylabel("Count", size=16)
        plt.show()
        print("-" * 100, "\n")
```
