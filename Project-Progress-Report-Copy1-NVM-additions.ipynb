### Abstract

### Introduction

Y2Y Harvard Square is a youth homeless shelter located in Cambridge, MA. It is one of only a handful of homeless shelters that track youth homelessness and services guests between the ages of 18 to 25. This segment of the homeless population is amongst the least studied homeless population. This lack of academic and government research hinders attempts for positive outcomes. 

Y2Y provides a comforting escape from the streets and provides accommodations, amenities and resources to assist guests in establishing an adequate foundation. Guests can either stay for one night or 30-days. This is determined through a lottery that is held every night slightly after mandatory check-in time. Those who stay are provided with resources, such as Youth Workers who build relationships and mentor guests, Case Managers who hold workshops and assist with goal achievement, students from Harvard Law who provide legal assistance, and physicians who provide medical care. 

The lack of research in this space  makes understanding the data that Y2Y collects vitally important. The goal of this project is to use this data to explore the factors that best predict guest satisfaction scores and guest outcomes.

### Datasets

The datasets utilized for the project consist of information recorded at the Y2Y facility by the Y2Y staff members and consist of guest surveys,  attendance, student case manager notes, youth worker notes, visiting professionals’ notes, such as physicians and Harvard Law students, and job/housing outcomes. That data provides insight on each guest during their stay. The datasets do not hold a large number of guests, since the program has only run for two full seasons, but each guest has a substantial amount of information recorded, yielding a large number of features to be associated for analysis. Each dataset and how we are using the data is described in detail below.

**Guest Survey** 

The Guest Survey dataset consists of survey results given to each 30-day/long-term stay guest upon their first or second night at Y2Y. Each survey includes demographic information, such as age, sex, and race, as well as other information, such as education history, sexual orientation, housing situation and job status. This information is crucial for understanding Y2Y’s guests and providing them with statistical inferences. The project utilizes this data to not only run statistical analysis but also to determine the prevalent features for predicting a guest’s satisfaction rating after their stay and whether they were able to find housing or a job. Ths data is extracted and combined with the exit survey data and cross checked with the other datasets to establish features for statistical analysis and classifier prediction. 

**Exit Survey**

The exit survey contains each 30-day/long-term guests’ responses prior to their leave. This dataset includes information like guest satisfaction scores, comments about future plans, and whether the guests rate themselves in a better situation after their stay at Y2Y. This information is the baseline for determining how Y2Y is performing and if Y2Y is making a substantial difference in their guests’ lives. Statistical analysis is done on this data to give Y2Y a clearer understanding of how they are rated. The data is also used in combination with features from the guest surveys and other datasets, to train the classifiers to how satisfied a guest will be based on their responses to the the guest survey and their attendance. 

**Attendance**

The attendance dataset is a list of each guest's’ attendance status for every night and whether they were present, called in, entered late, won an emergency 1-night lottery bed, or did not show up. This information is used to correlate guests’ presence to satisfactory and  consequences. 

**Lottery**

The Lottery dataset consists of the participants and outcomes of the daily lottery. This includes who entered, when they entered, whether they entered for a 1-night stay or a 30-day stay and if they won or lost. The data is used as an indicator for recurrence or 1-night or 30-day stay guests and correlated to the satisfaction and consequences data.  

**Consequences**

The Consequences dataset set lists the consequences recorded. The information includes the guest involved, the type of consequence such as rules broken, the severity, and the sanction. This data is used for correlation analysis with the Lottery dataset and satisfactory rating. 

**Season 2 Guest Housing & Jobs**

The Guest Housing & Jobs dataset includes the housing and employment status of some of the guests during season 2 of Y2Y’s opening. This includes whether they found a job or housing sometime during their stay or upon leaving. The information is used as ground truth indicators for determining if guests leave homelessness or unemployment during or after their stay at Y2Y.  

**Youth Worker Notes & Workshops**

The Youth Worker Notes & Workshops dataset outlines sporadic samples of notes from the Youth Worker sessions and workshops. Within these notes is what was discussed with certain guests, guests opinions and attendance. Since this data is mainly text and only relates to guest periodically by names in the text, it has several gaps. If time allows we will incorporate this data with the guest, exit, attendence, and housing/job data to rerun our models and anaylze the results.

**Completed**

Thus far the cleaning and merging of the top priority data has been completed. As one can see below in the implementation section we have run statiscal analysis on this data to get an optimal understanding of the data and further insight. This allows for deeper inference when running data analytical models on the data. 

First we ran statiscal analysis on the guest survey infromation. One feature we analyzed was the Age Distribution of Participant/Guests. The histogram shows that out of the range of 18 to 25 years old, the top 3 homeless youth that stayed for 30-day stays and filled out a guest survey were 24, 20, and 23 years old in that order. While the least number of guest were 18, 19 and 21. This could be an indicator of the time ranges when teenage youth becomes homeless, gets kicked out of their home, or voluntarily moves out. The ages of 24, 20, and 23 could be of significance when these young adults either graduate highschool or college and have no other place to go and could not find jobs. While on the otherhand the young adults who are 18, 19, or 21 could have jobs out of highschool and/or be in college during these ages but only be able to keep their, most likely non-career jobs, longer than a few years or end up graduating college with nowhere to go.

Another feature we ananlyzed is the Distribution of Guests and the zipcode/area they are from where they are from. From the analysis we can conclude that the most guests were from 02111. In Boston 02111 is the downtown area where many homeless conjugate to ask for food and money due to the high number of tourists. This area also holds several resource facilities and church organizations that provide help for the homeless. The analysis indicates that many of the Y2Y guest came from this area likely for these reasons. 

An additional feature we analyzed was the Distribution Types of Self Support that the guests relied on. Not surprisingly the majority of guests either relied on themselves with no additional support, or relied on themselves with unemployment benefits. This makes sense due to the fact that 90% of the guest are not only homeless but unemployed as well. 

Second we ran statiscal analysis on the Exit Survey information, specifically the satisfaction scores for the services that Y2Y provides. Overall the majority of guest were very satisfied with the safety, case managemnet support, advocacy and opportunities for leadership, legal resources, and medical services. Thus they would highly recommend Y2Y to friends and were satisfied with their overall experience at Y2Y. Although the guests's satifiaction with the food and feeling that they were in a better place after Y2Y were more evenly distributed and not as highly satisfied. They also felt similar about Y2Y helping them to better navigate service and systems for support and helping them plan a pathway out of homelessness. This could be indicators of opportunity for Y2Y to improve or incorporate their services that focus on assisting their guest to overcome homelessness and find a place to live. This may also include workshops to help them gain the mental strength and confidence to acheive their goals and work their way to stability.

We have completed a prelimary linear and logistic regression. We used cleaned and standardized guest survey data and safety feeling scores from the exit surveys in a multivarible linear regression. We used a 70% to 30% train test split. This was to determine the which features were most important in the guest feeling safe. Interestingly, the features that happened to have coefficents of the greatest magnitude were their age, if they fit the chronic definition of homelessnes, and if they had been seen a social worker. Looking at the correlation values it seems that adding more social worker visits would help improve satisfaction scores. Our next steps is to rerun this regression on more exit survey like guest satisfication and guest food satisifaction.

```python
def hist_exit_scores(df, column_score_names):
    for column in column_score_names:
        scores = df[column].dropna()
        print(column, "\n", "\n", df[column].describe(), "\n")
        bins=np.linspace(0, 9, 9, endpoint=False)
        # groups = resturaunt_rev_df.groupby(np.digitize(resturaunt_rev_df.stars, bins))
        fig, ax = plt.subplots(1,1, figsize=(12,9))
        ax.hist(scores, bins=bins, align='left', color="blue", rwidth=0.7)
        ax.set_xticks(bins)
        plt.title(column, size=16)
        plt.xlabel("Rating", size=16)
        plt.ylabel("Count", size=16)
        plt.show()
        print("-" * 100, "\n")
```

The Pearson correlation coefficient measures the linear relationship
 between two datasets. Strictly speaking, Pearson's correlation requires
 that each dataset be normally distributed. Like other correlation
 coefficients, this one varies between -1 and +1 with 0 implying no
 correlation. Correlations of -1 or +1 imply an exact linear
 relationship. Positive correlations imply that as x increases, so does
 y. Negative correlations imply that as x increases, y decreases.
 
 The p-value roughly indicates the probability of an uncorrelated system
 producing datasets that have a Pearson correlation at least as extreme
 as the one computed from these datasets. The p-values are not entirely
 reliable but are probably reasonable for datasets larger than 500 or so.
 
 Thus there is a 31.5% chacnce that if one increases the other increases.
